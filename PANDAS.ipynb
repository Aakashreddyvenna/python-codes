```python
import pandas as pd
```

```python
df=pd.read_csv("hotel_bookings.csv")
```

```python
df #printing the dataset
```

```python
'''dropna()=removes the row contain empty cells/null values
    or else u can change dropna(inplace=True) but once u use the dropna() it wont change to original one'''
df.dropna()
```

```python
#df.fillna(some value) is used to fill the empty cells or null values with some values
df.fillna(100)
```

```python
'''df[col name].fillna(val) is change for particular column
    or df[col name].mean/median/mode() we can apply for emoty cells '''
x=df["arrival_date_year"].mean()
df["arrival_date_year"].fillna(x)
```

```python
x=df["arrival_date_year"].median()
df["arrival_date_year"].fillna(x)
```

```python
x=df["arrival_date_week_number"].mode()
df["arrival_date_week_number"].fillna(x)
```
