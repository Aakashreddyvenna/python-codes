***MATPLOT***

```python
import matplotlib.pyplot as plt

x=[11,22,14,3,9]
y=[10,20,30,40,50]
plt.plot(x,y)
plt.title("Demo")
plt.xlabel("x-value")
plt.ylabel("y-value")
plt.show()
```

```python
lan=("c","python","java","c++")
users=(70,99,50,40)
plt.bar(lan,users,color="violet")
plt.title("programers useed lanuage")
plt.xlabel("lanuage")
plt.ylabel("users")
plt.show()
```

```python
x=[2,3,4,5,6,7,8,9]
plt.hist(x,bins=15,color="pink",edgecolor='black')
plt.xlabel("data")
plt.ylabel("frequency")
plt.title("histogram")
plt.show()
```

```python
x=[2,3,4,5,6,7,8,9]
y=[62,82,19,76,14,25,34,31]
plt.scatter(x,y,color='orange')
plt.xlabel("data")
plt.ylabel("y-data")
plt.title("scatter-plot")
```

```python
lan = ("c", "python", "java", "c++")
users = (70, 99, 50, 40)
plt.pie(users, labels=lan, autopct="%1.1f%%", colors=['red', 'blue', 'violet', 'pink'])
plt.title("pie-chart")
plt.show()
```

```python
data=(23,24,74,67,63,93,71,82,52)
plt.boxplot(data)
plt.title("Box-plot")
plt.show()
```

```python
lan=("c","python","java","c++")
users=(70,99,50,40)
plt.barh(lan,users,color="violet")
plt.title("programers useed lanuage")
plt.xlabel("lanuage")
plt.ylabel("users")
plt.show()
```

```python
lan=("c","python","java","c++")
users=(70,99,50,40)
plt.fill_between(lan,users,color="violet")
plt.title("area plot")
plt.xlabel("lanuage")
plt.ylabel("users")
plt.show()
```

```python
time=[1.6,2.44,3,4,5]
stock_price=[105,334,421,172,323]
plt.step(time,stock_price,color="green")
plt.title("step plot")
plt.xlabel("time")
plt.ylabel("stock_price")
plt.show()
```
